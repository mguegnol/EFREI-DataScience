This dataset includes 150 different flowers described by the **sizes** and **width** of two of their attributes: the **petals** and the **sepals**. These values are given in centimetres. <br/>
We can notice that the flowers present in this dataset are part of three different species: **Setosa**, **Versicolor** and **Virginica**. They are equitably distributed in this dataset: 50 records per species. <br>
It will be interesting to study which characteristic(s) **discriminates** each species and to what extent. <br>
To do this we will start by analyzing each species in relation to the others, in order to determine if one of the characters is not in the average. If so, we will analyse this character more precisely.



## is one of the parameter of the setosa specie far from the mean

```r

summarySetosa <- c(mean(iris[iris$Species=="setosa", "Sepal.Length"]),mean(iris[iris$Species=="setosa", "Sepal.Width"]),mean(iris[iris$Species=="setosa", "Petal.Length"]),mean(iris[iris$Species=="setosa", "Petal.Width"]))
summaryGlobal <- c(mean(iris[iris$Species!="setosa", "Sepal.Length"]),mean(iris[iris$Species!="setosa", "Sepal.Width"]),mean(iris[iris$Species!="setosa", "Petal.Length"]),mean(iris[iris$Species!="setosa", "Petal.Width"]))
compariason1 <- rbind(summarySetosa,summaryGlobal)

barplot(compariason1, 
        beside=T,
        names.arg=c("Sepal.Length","Sepal.Width","Petal.Length","Petal.Width"),
        ylim=c(0,6), 
        col=c("darkblue","red"))
legend("topright", 
       legend = c("Setosa", "not Setosa"), 
       fill = c("darkblue", "red"))
```

## We can see that the petal lenght's mean of the setosa is lower than the mean

## Let's look at what the distribution of the setosa's petal length look compare to the global distribution

```r
p1 <- hist(iris[iris$Species=="setosa", "Petal.Length"], plot=FALSE, breaks = 2)        
p2 <- hist(iris[,"Petal.Length"], plot=FALSE)

plot( p2, col=rgb(0,0,1,1/4),, xlim=c(0,10), ylim=c(0,40),main="Global") 
plot( p1, col=rgb(1,0,0,1/4),, xlim=c(0,10), ylim=c(0,40),main="Setosa")
```

```r
setosalt2 <- iris[(iris$Species=="setosa") & (iris$Petal.Length < 2.0), ]
globallt2 <- iris[(iris$Petal.Length < 2.0), ]
paste("proportion of petal between 2 and 5 cm from versicolor :", nrow(setosalt2)/nrow(globallt2)*100, "%")
```

## Setosa petal are the smallest, indeed all the petal which are under 2cm are from setosa
