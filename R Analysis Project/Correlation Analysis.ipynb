# Let's now focus on Correlation

```r
library(corrplot)
data("iris")
```

### First we want to see if a correlation exists between variables of all species.

```r
corrplot(cor(iris[c(1:4)]), type="upper", order="hclust", tl.col="black", tl.srt=45, title = "Correlation for All Species", mar=c(0,0,3,0))
```

### We can see that between all species, a clear correlation appears between Petals - Sepals and Petals - Petals

## Let's see those correlation for each Species

```r
corrplot(cor(iris[iris$Species == "setosa", 1:4]), type="upper", order="hclust", tl.col="black", tl.srt=45, title = "Correlation for Setosa Species", mar=c(0,0,3,0))
```

### Correlation for the Setosa Species show a proportionally between Speal L - Sepal W

```r
corrplot(cor(iris[iris$Species == "versicolor", 1:4]), type="upper", order="hclust", tl.col="black", tl.srt=45, title = "Correlation for Versicolor Species", mar=c(0,0,3,0))
```

### Versicolor Species shows a higer global correlation making it difficult to identify it's caractéristics. But we can see a significant correlation for  Petal L - Petal W and Sepal L - Petal L.

```r
corrplot(cor(iris[iris$Species == "virginica", 1:4]), type="upper", order="hclust", tl.col="black", tl.srt=45, title = "Correlation for Virginica Species", mar=c(0,0,3,0))
```

### Virginica only shows a high correlation for Sepal L - Petal L

## To illustrate those correlations we found, we will plot the dataset and try to visualize them

```r
plot(iris[iris$Species == "virginica", 1], iris[iris$Species == "virginica", 2], col="red", xlim=range(iris[1]), ylim=range(iris[2]), main="Relationship between Sepal Length and Width", xlab="Sepal Length", ylab = "Sepal Width")
points(iris[iris$Species == "setosa", 1], iris[iris$Species == "setosa", 2], col="blue")
points(iris[iris$Species == "versicolor", 1], iris[iris$Species == "versicolor", 2], col="green")
legend("topright", c("Virginica", "Versicolor", "Setosa"), col=c("red", "green", "blue"), cex=0.9, box.lty=1, pch=c(1))
```

```r
plot(iris[iris$Species == "virginica", 3], iris[iris$Species == "virginica", 4], col="red", xlim=range(iris[3]), ylim=range(iris[4]), main="Relationship between Petal Length and Width", xlab="Petal Length", ylab = "Petal Width")
points(iris[iris$Species == "setosa", 3], iris[iris$Species == "setosa", 4], col="blue")
points(iris[iris$Species == "versicolor", 3], iris[iris$Species == "versicolor", 4], col="green")
legend("topleft", c("Virginica", "Versicolor", "Setosa"), col=c("red", "green", "blue"), cex=0.9, box.lty=1, pch=c(1))
```

```r
plot(iris[iris$Species == "virginica", 3], iris[iris$Species == "virginica", 1], col="red", xlim=range(iris[3]), ylim=range(iris[1]), main="Relationship between Petal Length and Sepal Length", xlab="Petal Length", ylab = "Sepal Length")
points(iris[iris$Species == "setosa", 3], iris[iris$Species == "setosa", 1], col="blue")
points(iris[iris$Species == "versicolor", 3], iris[iris$Species == "versicolor", 1], col="green")
legend("topleft", c("Virginica", "Versicolor", "Setosa"), col=c("red", "green", "blue"), cex=0.9, box.lty=1, pch=c(1))
```

# The correlation we found shows a clear speparation on the graph, making those classes discriminative.
